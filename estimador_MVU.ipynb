```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
V = 12
sigma2 = 0.01
```

```python
def geraY(N,amostras):
  X = np.random.normal(V,np.sqrt(sigma2),(amostras,N))
  Y = np.mean(X,axis=1)
  return Y
```

```python
def experimento(N,rodadas):
  Y = geraY(N,rodadas)
  return np.mean(Y), np.var(Y)
```

```python
print(experimento(10,10000))
```

```python
x = geraY(1,100)
y = geraY(10,100)
plt.plot(x)
plt.plot(y,'r')
```

```python
Nmax = 100
rodadas = 1000
ymedio = np.zeros(Nmax)
yvar = np.zeros(Nmax)

for i in range(1,100):
  ymedio[i], yvar[i] =  experimento(i,rodadas)
```

```python
plt.plot(range(1,100),ymedio[range(1,100)])
```

```python
plt.plot(range(1,100),yvar[range(1,100)])
```
